# TASK 1: Exploratory Data Analysis (EDA)

## Objective
The objective of this task is to explore the dataset, understand its structure, identify patterns, detect anomalies, validate assumptions, and discover potential data quality issues before performing advanced analysis.

Dataset Used: Amazon Product Reviews Dataset

## TASK_1.1 - Ask Meaningful Questions.


```python
print("Question 1: What is the distribution of review scores?")
print("Question 2: Are positive reviews more common than negative reviews?")
print("Question 3: Are there any missing values in the dataset?")
print("Question 4: What is the average review rating?")
print("Question 5: Are there duplicate reviews present?")
```

## Task_1.2 - Explore Dataset Structure, Variables and Data Types.

```python
import pandas as pd

# Load Dataset
df = pd.read_csv("Reviews.csv")

print("Dataset Shape:", df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

df.head()
```

## Task_1.3 - Identify Trends, Patterns and Anomalies.

```python
print("Review Score Distribution:\n")
print(df["Score"].value_counts().sort_index())

print("\nAverage Rating:")
print(round(df["Score"].mean(), 2))
```

## Task_1.4 - Test Hypothesis and Validate Assumptions.

```python
positive_reviews = len(df[df["Score"] >= 4])
negative_reviews = len(df[df["Score"] <= 2])

print("Positive Reviews:", positive_reviews)
print("Negative Reviews:", negative_reviews)

if positive_reviews > negative_reviews:
    print("\nHypothesis Validated: Positive reviews are more common.")
else:
    print("\nHypothesis Rejected.")
```

## Task_1.5 - Detect Data Issues and Problems.

```python
print("Missing Values:\n")
print(df.isnull().sum())

print("\nDuplicate Records:")
print(df.duplicated().sum())
```
