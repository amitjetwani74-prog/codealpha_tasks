# TASK 2: Data Visualization

## Objective

The objective of this task is to transform raw review data into meaningful visualizations using Matplotlib and Seaborn. These visualizations help identify patterns, trends, and insights within the Amazon Product Reviews dataset.

## Task_2.1 - Import Required Libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_csv("Reviews.csv")

print("Dataset Loaded Successfully")
```

## Task_2.2 - Visualize Review Score Distribution.

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x=df["Score"],
    order=sorted(df["Score"].unique())
)

plt.title("Distribution of Review Scores")
plt.xlabel("Review Score")
plt.ylabel("Number of Reviews")

plt.show()
```

## Task_2.3 - Visualize Positive vs Negative Reviews.

```python
positive_reviews = len(df[df["Score"] >= 4])
negative_reviews = len(df[df["Score"] <= 2])

plt.figure(figsize=(6,6))

plt.pie(
    [positive_reviews, negative_reviews],
    labels=["Positive Reviews", "Negative Reviews"],
    autopct="%1.1f%%"
)

plt.title("Positive vs Negative Reviews")

plt.show()
```

## Task_2.4 - Visualize Top Review Scores.

```python
score_counts = df["Score"].value_counts().sort_index()

plt.figure(figsize=(8,5))

plt.bar(
    score_counts.index,
    score_counts.values
)

plt.title("Review Score Frequency")
plt.xlabel("Score")
plt.ylabel("Count")

plt.show()
```

## Task_2.5 - Visualize Review Length Distribution.

```python
df["Review_Length"] = df["Text"].astype(str).apply(len)

plt.figure(figsize=(8,5))

sns.histplot(
    df["Review_Length"],
    bins=30
)

plt.title("Distribution of Review Length")
plt.xlabel("Review Length")
plt.ylabel("Frequency")

plt.show()
```

## Task_2.6 - Correlation Between Helpfulness and Score.

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x=df["HelpfulnessNumerator"],
    y=df["Score"]
)

plt.title("Helpfulness vs Review Score")
plt.xlabel("Helpfulness")
plt.ylabel("Score")

plt.show()
```
