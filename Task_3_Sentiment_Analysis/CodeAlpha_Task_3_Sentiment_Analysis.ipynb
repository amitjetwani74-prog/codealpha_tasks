# TASK 3: Sentiment Analysis

## Objective

The objective of this task is to analyze customer review text and classify sentiments as Positive, Neutral, or Negative. This helps understand customer opinions, product perception, and overall satisfaction.

## Task_3.1 - Load Dataset and Create Sentiment Labels.

```python
import pandas as pd

df = pd.read_csv("Reviews.csv")

def get_sentiment(score):
    if score >= 4:
        return "Positive"
    elif score == 3:
        return "Neutral"
    else:
        return "Negative"

df["Sentiment"] = df["Score"].apply(get_sentiment)

df[["Score","Sentiment"]].head()
```

## Task_3.2 - Analyze Sentiment Distribution.

```python
sentiment_counts = df["Sentiment"].value_counts()

print("Sentiment Distribution:\n")
print(sentiment_counts)
```

## Task_3.3 - Visualize Positive, Neutral and Negative Reviews.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.countplot(
    x=df["Sentiment"],
    order=["Positive","Neutral","Negative"]
)

plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")

plt.show()
```

## Task_3.4 - Analyze Public Opinion and Trends.

```python
total_reviews = len(df)

positive_percent = (len(df[df["Sentiment"]=="Positive"]) / total_reviews) * 100
neutral_percent = (len(df[df["Sentiment"]=="Neutral"]) / total_reviews) * 100
negative_percent = (len(df[df["Sentiment"]=="Negative"]) / total_reviews) * 100

print(f"Positive Reviews: {positive_percent:.2f}%")
print(f"Neutral Reviews: {neutral_percent:.2f}%")
print(f"Negative Reviews: {negative_percent:.2f}%")
```

## Task_3.5 - Business Insights from Sentiment Analysis.

```python
positive = len(df[df["Sentiment"]=="Positive"])
negative = len(df[df["Sentiment"]=="Negative"])

if positive > negative:
    print("Most customers are satisfied with the products.")
    print("Products have a positive market perception.")
    print("Businesses can use this feedback for marketing and promotions.")
else:
    print("Products may require improvement based on customer feedback.")
```
